```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
output = pd.read_csv("output.csv", index_col=0)
```

```python
output["error_1"] =  output["dl_mkt_price"] - output["ml_mkt_price"]
output["error_2"] =  output["dl_mkt_price"] - output["ml_learned_price"]
```

```python
output
```

```python
output.describe()
```

```python
worst_option = output.index[output.error_1.argmin()]
best_option = output.index[output.error_1.argmax()]

worst_option_df = pd.read_csv(f"data/{worst_option}.csv", index_col=0)
best_option_df = pd.read_csv(f"data/{best_option}.csv", index_col=0)
```

```python
worst_option_df.S.plot()
plt.axhline(y=worst_option_df["strike_price"].iloc[0] / 1000, color='r', linestyle='-')
```

```python
best_option_df.S.plot()
plt.axhline(y=best_option_df["strike_price"].iloc[0] / 1000, color='r', linestyle='-')
```
